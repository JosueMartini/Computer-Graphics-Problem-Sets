### Imports and Libraries

```python
import numpy as np
import math
from PIL import Image
from matplotlib import pyplot as plt
```

### Global Variables

```python
path = "C:/Users/josue/OneDrive/Desktop/Images/"
img1 = Image.open(path + "img1.jpg")
img2 = Image.open(path + "img2.jpg")
img3 = Image.open(path + "img3.jpg")
img4 = Image.open(path + "img4.jpg")
img5 = Image.open(path + "img5.jpg")
```

### Image 1

```python
# code and method definitions
array1 = np.asarray(img1)
yrows, xcols = array1.shape
points = np.mgrid[0:xcols, 0:yrows]
points = points.reshape( ( 2, yrows * xcols) )

plt.imshow(array1, cmap = "gray")

p1x = 75
p1y = 151

plt.plot(p1x,p1y, marker = ".", color = "cyan", markersize = 10)

p2x = 183
p2y = 207

plt.plot(p2x,p2y, marker = ".", color = "lime", markersize = 10)

p3x = 40
p3y = 220

plt.plot(p3x,p3y, marker = ".", color = "red", markersize = 10)

hyp1 = np.sqrt( (p1x - p2x)**2 + (p1y - p2y)**2 )

a = (p2x - p1x) / hyp1
c = (p2y - p1y) / hyp1

hyp2 = np.sqrt( (p1x - p3x)**2 + (p1y - p3y)**2 )

b = (p3x - p1x) / hyp2
d = (p3y - p1y) / hyp2

M = np.array( [ [a,b], [c,d] ] )

pprime = M @ points

pprime = pprime.round().astype(int)

xoffset = p1x
yoffset = p1y

pprime[0] += xoffset
pprime[1] += yoffset

pprime = pprime.reshape( (2, yrows, xcols), order= "F" )


```

```python
# test cases and output (be sure to include all test cases as indicated)
xmat, ymat = pprime.reshape( (2, yrows, xcols), order="F" )
indexvals = xmat + xcols * ymat
newimg1 = np.take(array1, indexvals, mode="wrap")

nrows = 1
ncols = 3
fig, axis = plt.subplots(nrows, ncols)
axis[0].imshow(img1,cmap="gray")
axis[1].imshow(newimg1, cmap="gray")

height = 80
width = 123

newimg1 = newimg1[ 0:int(height) , 0:int(width)  ]
axis[2].imshow(newimg1, cmap="gray")
```
