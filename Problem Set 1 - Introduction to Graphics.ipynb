### Im

```python
# code and method definitions
def get_stats( img ) :

    sum = 0
    sum2 = 0
    count = 0
    
    
    array1 = np.asarray(img)
    for row in array1:
        for col in row:
            sum = sum + col
            count = count + 1
    mean = sum / count

    for row in array1:
        for col in row:
            square = pow( ( col - mean ), 2)
            sum2 = sum2 + square
    var = sum2 / count
    sd = math.sqrt(var)
    return mean, var, sd
    
```

```python
# test cases and output (be sure to include all test cases as indicated)
mean, var, sd = get_stats( img1 )
print( f"Mean:  {mean}    Variance: {var}     Standard Deviation:  {sd}"  )


mean, var, sd = get_stats( img2 )
print( f"Mean:  {mean}    Variance: {var}     Standard Deviation:  {sd}"  )


mean, var, sd = get_stats( img3 )
print( f"Mean:  {mean}    Variance: {var}     Standard Deviation:  {sd}"  )
```

### Problem 2: get_lighter_image()

```python
# code and method definitions
def get_lighter_image( img, int ):
    
    nrows = 1
    ncols = 2
    fig, axis = plt.subplots(nrows, ncols)
    axis[0].imshow(img,cmap="gray")

    array1 = np.array(img)
     
    for row in range(array1.shape[0]):
        for col in range(array1.shape[1]):
            array1[row, col] = np.minimum(array1[row, col] + int, 255)
    return axis[1].imshow(array1, cmap = "gray")

  


    
    
```

```python
# test cases and output (be sure to include all test cases as indicated)
get_lighter_image( img1, 30 )
```

### Problem 3: get_mirror()

```python
# code and method definitions
def get_mirror( img, k ):

    nrows = 1
    ncols = 2
    fig, axis = plt.subplots(nrows, ncols)
    axis[0].imshow(img,cmap="gray")
    oldImg = np.array(img)
    height, width = img.size[::-1]
    newImg = np.empty((height, width))

    if k >= width // 2:
        for row in range(height):
            for col in range(k, width):
                newImg[row, col] = oldImg[row, col-k]

        for row in range(height):
            for col in range(k):
                newImg[row, col] = oldImg[row, k - 1 - col]
    else:
        for row in range(height):
            for col in range(k, width):
                newImg[row, k - (col - k) - 1] = oldImg[row, col]
                
        for row in range(height):
            for col in range(k, width):
                newImg[row, col] = oldImg[row, col]
            
    axis[1].imshow(newImg, cmap= "gray")
    
            
```

```python
# test cases and output (be sure to include all test cases as indicated)
get_mirror(img2, 200)
```

### Problem 4: get_border()

```python
# code and method definitions
def get_border(img, k):
    
    nrows = 1
    ncols = 2
    fig, axis = plt.subplots(nrows, ncols)
    axis[0].imshow(img,cmap="gray")

    w = int(k * img.width)

    width, height = img.size

    newImg = img.copy()

    for row in range(w):
        for col in range(height):
            newImg.putpixel((row, col), 0)
            newImg.putpixel((width - 1 - row, col), 0)

    for row in range(width):
        for col in range(w):
            newImg.putpixel((row, col), 0)
            newImg.putpixel((row, height - 1 - col), 0)
    
    axis[1].imshow(newImg, cmap="gray")
    
```

```python
# test cases and output (be sure to include all test cases as indicated)
get_border( img1, 0.15 )
```
